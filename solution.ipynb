```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df_data_set = pd.read_csv("drinkMenu.csv")
```

```python
df_data_set.iloc[237,4]=3.2

df_data_set.isnull().sum()
```

```python
# df_data_set[' Total Fat (g)']=df_data_set[' Total Fat (g)'].to_numeric()
df_data_set[' Total Fat (g)']=df_data_set[' Total Fat (g)'].astype('double')
```

```python
df_data_set.columns
```

```python
newdf_data_set = df_data_set.drop_duplicates()
newdf_data_set['Beverage_category'].fillna(newdf_data_set['Beverage_category'].mode()[0], inplace=True)
newdf_data_set['Beverage'].fillna(newdf_data_set['Beverage'].mode()[0], inplace=True)
newdf_data_set['Beverage_prep'].fillna(newdf_data_set['Beverage_prep'].mode()[0], inplace=True)
newdf_data_set['Calories'].fillna(newdf_data_set['Calories'].mode()[0], inplace=True)
newdf_data_set[' Total Fat (g)'].fillna(newdf_data_set[' Total Fat (g)'].mode()[0], inplace=True)
newdf_data_set['Trans Fat (g) '].fillna(newdf_data_set['Trans Fat (g) '].mode()[0], inplace=True)
newdf_data_set['Saturated Fat (g)'].fillna(newdf_data_set['Saturated Fat (g)'].mode()[0], inplace=True)
newdf_data_set[' Sodium (mg)'].fillna(newdf_data_set[' Sodium (mg)'].mode()[0], inplace=True)
newdf_data_set[' Total Carbohydrates (g) '].fillna(newdf_data_set[' Total Carbohydrates (g) '].mode()[0], inplace=True)
newdf_data_set['Cholesterol (mg)'].fillna(newdf_data_set['Cholesterol (mg)'].mode()[0], inplace=True)
newdf_data_set[' Dietary Fibre (g)'].fillna(newdf_data_set[' Dietary Fibre (g)'].mode()[0], inplace=True)
newdf_data_set[' Sugars (g)'].fillna(newdf_data_set[' Sugars (g)'].mode()[0], inplace=True)
newdf_data_set[' Protein (g) '].fillna(newdf_data_set[' Protein (g) '].mode()[0], inplace=True)
newdf_data_set['Vitamin A (% DV) '].fillna(newdf_data_set['Vitamin A (% DV) '].mode()[0], inplace=True)
newdf_data_set['Vitamin C (% DV)'].fillna(newdf_data_set['Vitamin C (% DV)'].mode()[0], inplace=True)
newdf_data_set[' Calcium (% DV) '].fillna(newdf_data_set[' Calcium (% DV) '].mode()[0], inplace=True)
newdf_data_set['Iron (% DV) '].fillna(newdf_data_set['Iron (% DV) '].mode()[0], inplace=True)
newdf_data_set['Caffeine (mg)'].fillna(newdf_data_set['Caffeine (mg)'].mode()[0], inplace=True)

```
